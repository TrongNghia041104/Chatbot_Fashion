# Research Demo v3 - Split Notebook Index

Folder này chứa các notebook nhỏ được tách từ notebook gốc:

`Chatbot_RAG_MultiModal_ViFashionCLIP_research_demo_v3.ipynb`

Mục tiêu: đọc hiểu và debug từng tầng của hệ RAG thay vì chạy một notebook khổng lồ.

## Thứ tự nên đọc/chạy

1. `01_environment_config_models.ipynb` - Environment, Config, Models
2. `02_product_data_pipeline.ipynb` - Product Data Pipeline
3. `03_layer_a_text_retrieval_debug.ipynb` - Layer A Text Retrieval Debug
4. `04_image_retrieval_debug.ipynb` - Image Retrieval Debug
5. `05_layer_b_outfit_and_category_mapping.ipynb` - Layer B Outfit And Category Mapping
6. `06_vision_module_debug.ipynb` - Vision Module Debug
7. `07_llm_prompts_chains_history.ipynb` - LLM, Prompts, Chains, History
8. `08_router_security_logging_eval.ipynb` - Router, Security, Logging, Eval
9. `09_chat_loop_demo.ipynb` - Full Chat Loop Demo

## Nguyên tắc debug

1. Không bắt đầu từ chat loop.
2. Kiểm tra service/config trước.
3. Debug raw retrieval trước khi sửa prompt.
4. Với câu phối đồ, kiểm tra Layer B rule trước rồi mới kiểm tra sản phẩm Layer A.
5. Với lỗi hallucination, xem final docs gửi vào LLM trước.

## Ghi chú

- Notebook gốc không bị sửa/xoá.
- Output trong các notebook tách đã được clear để tránh nhìn nhầm kết quả cũ.
- Nếu một notebook có phụ thuộc, phần đầu notebook đã ghi rõ nên chạy notebook nào trước.
